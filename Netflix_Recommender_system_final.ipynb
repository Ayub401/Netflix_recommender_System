```python
# convert the str to list: removing the comma
import ast
ast.literal_eval("[1,2,3,4]")
```

```python
# extracting name key from each dic
def Convert(obj):
    Name_lst = []
    for i in ast.literal_eval(obj):  # covert the str to list.
        Name_lst.append(i["name"])       # extracting the value of name key.
    return Name_lst   
```

```python
# from cast we need to extract name of top 3 actor of the movie.

# extracting name key from each dic
def Convert_Cast(obj):
    Top_3_actor_name = []
    Counter = 0
    for i in ast.literal_eval(obj):
        if Counter !=3:
            Top_3_actor_name.append(i["name"])
            Counter +=1
        else:
            break
    return Top_3_actor_name   
```

```python
# need to apply stemming
from nltk.stem.porter import PorterStemmer
Stemmer = PorterStemmer()


# note PorterStemmer takes list of words as input.
```

```python
def stemming(text):
    lst_of_stem_words = []
    for i in text.split():  # i is the lst words in a tag
        lst_of_stem_words.append(Stemmer.stem(i))
    return " ".join(lst_of_stem_words)     # again converting the lst of words to string
```

```python
New_df["tags"] = New_df["tags"].apply(stemming)
```

```python
New_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
TF_idf = TfidfVectorizer(max_features= 5000, stop_words= "english")


# it will create a vector of 5000 words.
#the words will be selected on the basis of the thier freuancy
# less frequency word will be ignored
# english stop wrods will be removed.
```

```python
Vector = TF_idf.fit_transform(New_df["tags"]).toarray()
```

```python
# we will use cosine similary to select the simalr movies.
from sklearn.metrics.pairwise import cosine_similarity
Similarity = cosine_similarity(Vector)
```

```python
# making a function to extract similar movie based on similarity

def Recommendation(movie):
    movie_index = New_df[New_df["title"] == movie].index[0]   # extractin the index of a movie
    Distance = Similarity[movie_index]  # extracting the similary of that movie with other movie based on the index
    Top_5_movies = sorted(list(enumerate(Distance)), reverse = True, key = lambda x:x[1])[1:6]  # extracting the top five movies based on the distance
    
    #the top_5_movies will contain tuple of movies id and distace 0, 1.0000000000000002) check above
    for tup_movie in Top_5_movies:
        print(tup_movie[0])
```
